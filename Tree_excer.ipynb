###### Solution

```python
class TreeNode:
    def __init__(self, data):
        # assert(type(data) == tuple)
        self.data = data
        self.children = []
        self.parent = None

    def get_level(self):
        level = 0
        p = self.parent
        while p:
            level += 1
            p = p.parent

        return level

    def add_child(self, child):
        child.parent = self
        self.children.append(child) #child must be treenode object


    def print_tree(self, type=None):
        self.type = type
        spaces = ' ' * self.get_level() * 3
        prefix = spaces + "|__" if self.parent else ""
        
        if self.type == 'name':
            value = self.data.split(' (')[0]

        elif self.type == 'title':
            value = self.data.split(' (')[1][:-1]

        elif self.type == 'both':
            value = self.data

        print(prefix + value)
        if self.children: 
            for child in self.children:
                child.print_tree(self.type)




def build_management_tree():
    infra_head = TreeNode('Vishwa (Infrastructure Head)')
    infra_head.add_child(TreeNode('Dhaval (Cloud Manager)'))
    infra_head.add_child(TreeNode('Abhijit (App Manager)'))
    app_head = TreeNode('Aamir (Application head)')
    
    hr_head = TreeNode('Gels (HR Head)')
    hr_head.add_child(TreeNode('Peter (Recruitment Manager)'))
    hr_head.add_child(TreeNode('Waqas (Policy Manager)'))

    ceo = TreeNode('Nilupul (CEO)')
    cto = TreeNode('Chinmay (CTO)')
    
    cto.add_child(infra_head)
    cto.add_child(app_head)

    ceo.add_child(cto)
    ceo.add_child(hr_head)
    
    return ceo


if __name__ == '__main__':
    mgt_node = build_management_tree()
    mgt_node.print_tree('name')
    print('\n')
    mgt_node.print_tree('title')
    print('\n')
    mgt_node.print_tree('both')
```

[Q2](https://github.com/codebasics/data-structures-algorithms-python/blob/master/data_structures/7_Tree/7_tree_exercise.md)
###### Solution

```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = []
        self.parent = None

    def get_level(self):
        level = 0
        p = self.parent
        while p:
            level += 1
            p = p.parent

        return level

    def print_tree(self, level=0):
        self.level = level
        spaces = ' ' * self.get_level() * 3
        prefix = spaces + "|__" if self.parent else ""
        print(prefix + self.data)

        if self.children:
            for child in self.children:
                if child.get_level() <= self.level:
                    child.print_tree(self.level)
                elif self.level == 0:
                    child.print_tree()

    def add_child(self, child):
        child.parent = self
        self.children.append(child)


def build_locaton_tree():
    abia = TreeNode('Abia State')
    abia.add_child(TreeNode('Abriba'))

    ngr = TreeNode('Nigeria')  
    ngr.add_child(abia)
    ngr.add_child(TreeNode('Lagos State'))
    ngr.add_child(TreeNode('Rivers State'))

    new_jersey = TreeNode('New Jersey')
    new_jersey.add_child(TreeNode('Princeton'))
    new_jersey.add_child(TreeNode('Trenton'))

    california = TreeNode('California')
    california.add_child(TreeNode('San Francisco'))
    california.add_child(TreeNode('Mountian View'))
    california.add_child(TreeNode('Palo Alto'))

    usa = TreeNode('USA')
    usa.add_child(new_jersey)
    usa.add_child(california)

    world = TreeNode('Global')
    world.add_child(usa)
    world.add_child(ngr)

    return world

root = build_locaton_tree()
root.print_tree()
root.print_tree(1)
root.print_tree(2)
root.print_tree(3)

```
